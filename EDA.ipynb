```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('World_Populations.csv')

df
```

```python
# set the float format to display two decimal places
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# info of the dataset
df.info()
```

```python
# descriptive statistics of the dataset using describe()
df.describe()
```

```python
# Datatypes for each column:
df.dtypes
```

```python
df.select_dtypes(include='number')
```

```python
df.select_dtypes(include='object')
```

```python
# Counting of all null values in the dataset
df.isnull().sum()
```

```python
# Count of unique values in each column
df.nunique()
```

```python
# Sorting on values
df.sort_values(by="2022 Population", ascending=False).head(10) # by default, sort is done in ascending order
```

```python
# Correlation between columns:
df.corr(numeric_only=True)
```

```python
# display the correlation matrix as a heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.rcParams['figure.figsize']=(20,7)
plt.show()
```

```python
# Checking the data for Oceania
df[df['Continent'].str.contains('Oceania')]
```

```python
# Grouping data by this usecase "Are there certain continents that are growing faster than others? And in which ways?"

df.groupby('Continent').mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
```

```python
# Visualization for the usecase 
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
df2
```

```python
df2.plot()
```

```python
df3 = df2.transpose()
df3
```

```python
df3.plot()
```

```python
# Boxplot for outliers
df.boxplot(figsize=(20,10))
```
